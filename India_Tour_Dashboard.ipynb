```python
def elemination(df, cities):
    #print('cities',cities)
    a = ['MUMBAI', 'DELHI', 'BANGLORE']
    newlist = []
    while(True):
        choice = input('\nEnter City You do not wish to visit: ')
        if choice in a:
            print('\nYou cannot Eliminate "MUMBAI", "DELHI" or "BANGLORE".')
        elif choice in cities:
            #print('\nChoice in Cities')
            newlist.append(choice)
            break
        else:
            print('\nInvalid INPUT')

    while(True):
        choice = input('\nEnter Second City You do not wish to visit: ')
        if choice in a:
            print('\nYou cannot Eliminate "MUMBAI", "DELHI" or "BANGLORE".')
        elif choice in cities:
            #print('\nChoice in Cities')
            newlist.append(choice)
            break
        else:
            print('\nInvalid INPUT')

    while(True):
        choice = input('\nEnter Third City You do not wish to visit: ')
        if choice in a:
            print('\nYou cannot Eliminate "MUMBAI", "DELHI" or "BANGLORE".')
        elif choice in cities:
            #print('\nChoice in Cities')
            newlist.append(choice)
            break
        else:
            print('\nInvalid INPUT')
            
    #print(newlist)
    finallist = []
    for i in cities:
        if i in newlist:
            continue
        else:
            finallist.append(i)
    return finallist
```
